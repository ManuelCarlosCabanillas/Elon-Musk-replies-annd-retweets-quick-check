**Look for tweets that are replies or are to specific people**

```python
import numpy as np

who_he_replies_to =[]
for tweet in em_tweets.text_decode:
    if (tweet[:1]=='@'):
        # Itss a reply
        parts = tweet.split()
        for i in range (0, len(parts)):
            if (parts[i][0]=='@'):
                who_he_replies_to.append(parts[i])
```
